##### Tách đuôi email để phân loại tài khoản truy cập

```python
a = df['Email'].str.split('@', expand=True)
a[1].value_counts()
```

```python
df['Type_email'] = np.select(
    [
        df['Email'].str.contains('@.*edu', case=False, na=False),
        df['Email'].str.contains('uniace', case=False, na=False),
        df['Email'].str.contains('@.*gmail', case=False, na=False)
    ],
    [
        'Sinh viên',
        'Thành viên',
        'Gmail'
    ],
    default='Khác'
)
df.head(5)
```

```python
b = df['ma_path'].str.split('/', expand=True)
df['path1'] = b[1]
```

##### Chuẩn hóa lại tên cột

```python
# safer: rename bya column names (works regardless of current columnd order or count)
df.rename(
	columns={
		'Email': 'email',
		'Type': 'type',
		'Name': 'name',
		'Title': 'title',
		'MA URL': 'ma_url',
		'MA Referrer': 'ma_referrer',
		'ma_path': 'ma_path',
		'IP Address': 'ip_address',
		'cuid': 'cuid',
		'Date': 'datetime',
		'Date_only': 'date',
		'Time': 'time',
		'Type_email': 'email_type',
        'path1': 'path1'
	},
	inplace=True
)
df.sample(5)
```
